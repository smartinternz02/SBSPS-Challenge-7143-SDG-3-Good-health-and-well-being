# Hypertension among Adults (age 15 years and above) Men And Women


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas_profiling as pp
```

```python
#importing the file and copying the data to another variable
df=pd.read_csv('hackathon2.csv')
df_copy=df.copy()
df.head()
```

```python
#Assigning the rows from 91-96 of the choosen category(Hypertension among Adults (age 15 years and above) Women)

df_new=df.loc[91:96,:]
df_new
```

# Cleaning the data


```python
#In the choosen category there are few null values and they are done manually in excel sheets,coversions of data types are done.
```

```python
#Giving a new index for the data
k=0
for i in range(91,97):
    df_new.loc[i,'Sl.No']=k
    k=k+1
df_new
```

```python
df_new=df_new.set_index('Sl.No')
df_new
```

```python
#Removing the column as the whole column has null values(Process Of Cleaning)
df_new.drop(columns=['Total (2015-16)'],inplace=True)
```

```python
df_new
```

```python
df_new.dtypes
```

```python
#The given dataset columns are in the object form ,converting to numeric ,So that  data normalization is done corerctly
for col in list(df_new.columns)[2:]:
    df_new[col] = pd.to_numeric(df_new[col],errors='ignore')
df_new
```
